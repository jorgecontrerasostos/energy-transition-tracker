```python
import pandas as pd
import duckdb
file = 's3://energy-transition-tracker/yearly_full_release_long_format.csv'
```

```python
df = pd.read_csv(file)
df
```

```python
df.columns
```

```python
df['Area'].unique()
```

```python
mexico_data = df[df['Area'] == 'Panama']
mexico_data
```

```python
def format_columns(column: str) -> str:
    column = column.lower().split(' ')
    return '_'.join(column)

mexico_data.columns = [format_columns(column) for column in mexico_data.columns]
mexico_data.head()
```

```python
mexico_data.drop(['iso_3_code', 'area_type', 'continent', 'ember_region', 'eu', 'oecd', 'g20', 'g7', 'asean'], axis=1, inplace=True)
```

```python
mexico_data
```

```python
cols = ['category', 'subcategory', 'variable', 'unit']
for col in mexico_data.columns:
    if col in cols:
        print(mexico_data[col].name)
        print(mexico_data[col].unique())
```
